#####Ejercicio 1: Calcular y graficar las medidas de tendencia central, localización y forma, segmentadas por rango_salarial ¿Qué podemos concluir para cada una de las escalas salariales analizadas?


```python
#Segmentacion Bajo
df_bajo = df.loc[df['rango_salarial'] == 'bajo']
df_bajo
```

```python
#Segmentacion medio
df_medio = df.loc[df['rango_salarial'] == 'medio']
df_medio
```

```python
#Segmentacion alto
df_alto = df.loc[df['rango_salarial'] == 'alto']
df_alto
```

```python
mediana=df_bajo['Salary'].median()
media=df_bajo['Salary'].mean()
moda=max(df_bajo['Salary'].mode())
q1 = df_bajo['Salary'].quantile(0.25)
q3 = df_bajo['Salary'].quantile(0.75)
sns.histplot(data=df_bajo,x='Salary',kde=True)
plt.rcParams['figure.figsize'] = (8,6)
sns.set_style("darkgrid")
sns.kdeplot(df_bajo['Salary'])
plt.plot((media,media),(0,2000),color = "red",label='Media')
plt.plot((mediana,mediana),(0,2000),color = "yellow",label='Mediana')
plt.plot((moda,moda),(0,2000),color="black",label='Moda')
plt.plot((q1,q1),(0,2000),color = "#ff01f0",label='Q1')
plt.plot((q3,q3),(0,1000),color = "#ff01f0",label='Q3')
plt.title('Salario de las empresas de Software: Segmento Bajo')
plt.xlabel('Salary')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.legend()
plt.show()
print(media)
print(mediana)
print(moda)
print(q1)
print(q3)
```

**Análisis** El gráfico representa la variable 'Salary', específicamente para el segmento 'bajo' de los datos. Los salarios tienden a estar agrupados cerca de la mediana, con una ligera simetría hacia la derecha debido a la presencia de algunos valores más altos y atípicos. La moda se encuentra en un punto bajo, lo que sugiere una concentración de salarios en los valores más bajos del rango, pero también muchos salarios por encima de la moda. La diferencia entre el tercer cuartil (Q3) y el primer cuartil (Q1) nos da una idea de la dispersión de los datos



```python
#segmento medio
mediana_medio=df_medio['Salary'].median()
media_medio=df_medio['Salary'].mean()
moda_medio=max(df_medio['Salary'].mode())
q1_medio = df_medio['Salary'].quantile(0.25)
q3_medio = df_medio['Salary'].quantile(0.75)
sns.histplot(data=df_medio,x='Salary',kde=True)
plt.rcParams['figure.figsize'] = (8,6)
sns.set_style("darkgrid")
sns.kdeplot(df_medio['Salary'])
plt.plot((media_medio,media_medio),(0,2000),color = "red",label='Media')
plt.plot((mediana_medio,mediana_medio),(0,2000),color = "yellow",label='Mediana')
plt.plot((moda_medio,moda_medio),(0,2000),color="black",label='Moda')
plt.plot((q1_medio,q1_medio),(0,2000),color = "#ff01f0",label='Q1')
plt.plot((q3_medio,q3_medio),(0,1000),color = "#ff01f0",label='Q3')
plt.title('Salario de las empresas de Software: Segmento Medio')
plt.xlabel('Salary')
plt.ylabel('Densidad')
plt.grid(True)
plt.legend()
plt.show()
print(q1_medio)
print(q3_medio)
print(moda_medio,media_medio,mediana_medio)
```

**Análisis** El gráfico representa la variable 'Salary', específicamente para el segmento 'medio' de los datos. Se puede inferir que la media de los salarios son relativamente altos en este segmento. La asimetría positiva observada a la derecha indica que hay una mayor concentración de salarios en el rango inferior, pero también hay salarios mucho más altos que están en tendencia del promedio hacia arriba. Los múltiples picos en la distribución sugieren que puede haber diferentes grupos salariales dentro de este segmento medio.

```python
#segmento alto
mediana=df_alto['Salary'].median()
media=df_alto['Salary'].mean()
moda=max(df_alto['Salary'].mode())
q1 = df_alto['Salary'].quantile(0.25)
q3 = df_alto['Salary'].quantile(0.75)
sns.histplot(data=df_alto,x='Salary',kde=True)
plt.rcParams['figure.figsize'] = (8,6)
sns.set_style("darkgrid")
sns.kdeplot(df_alto['Salary'])
plt.plot((media,media),(0,2000),color = "red",label='Media')
plt.plot((mediana,mediana),(0,2000),color = "yellow",label='Mediana')
plt.plot((moda,moda),(0,2000),color="black",label='Moda')
plt.plot((q1,q1),(0,2000),color = "#ff01f0",label='Q1')
plt.plot((q3,q3),(0,1000),color = "#ff01f0",label='Q3')
plt.title('Salario de las empresas de Software: Segmento Alto')
plt.xlabel('Salary')
plt.ylabel('Densidad')
plt.grid(True)
plt.legend()
plt.show()
print(q1)
print(q3)
print(moda,media,mediana)
```

**Análisis:** El gráfico representa la variable 'Salary', específicamente para el segmento 'alto' de los datos. Permitiendo inferir que, aunque hay algunas empresas que ofrecen salarios altos, la mayoría de las empresas en este segmento tienen salarios bajos, y las medidas de tendencia central no reflejan de manera precisa la realidad salarial del segmento alto. Finalmente la Media, Mediana y Moda: Todas están muy cerca de 0, lo que puede indicar la presencia de valores atípicos que distorsionan la distribución. Esto significa que, aunque la mayoría de los salarios son bajos, hay algunas empresas que pagan salarios muy altos, lo que afecta el promedio. Q1 y Q3: También están cerca de 0, reforzando la idea de que la mayoría de los salarios están en el extremo inferior de la escala.
